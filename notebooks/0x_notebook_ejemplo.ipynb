# Introducción a los notebook de Jupyter





Un notebook es un hiperdocumento que combina código y texto enriquecido. Permite tener ecuaciones, figuras, videos, animaciones, todo en un mismo documento. Esto facilita tener documentos que permitan presentar elementos conceptuales en los que se incluten pequeñas porciones de código (_snippets_) que se ejecutan de forma interactiva.

En el curso de Modelación computacional usaremos los notebooks como material de apoyo.

## Celdas de texto


Los notebooks están divididos en celdas, las cuales tienen 

```ipython
In [ ]:
```

al lado izquierdo. Hay dos tipos básicos de celdas:

- Celdas de texto; y
- Celdas de código.


**Esta es una celda de texto** y como tal permite:

- Incluir gráficas (un gif en este caso).


![alt text](https://upload.wikimedia.org/wikipedia/commons/e/e3/Catenoid.gif "Ejemplo de una catenoide.")

- Incluir ecuaciones usando sintaxis de $\LaTeX$ a través de [MathJax](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Typesetting%20Equations.html). Por ejemplo, el código

```latex
$$E = mc^2$$
```

genera la siguiente ecuación.

$$E = mc^2$$

- También permite *dar* formato al **texto**.

- E incluir [hipervínculos](https://en.wikipedia.org/wiki/Euler%E2%80%93Bernoulli_beam_theory).


## Celdas de código



Las celdas de código permiten ejecutar bloques de código.

Es común (y buena práctica) realizar el llamado a los módulos o bibliotecas al inicio del código.

- Las bibliotecas incluyen código para realizar tareas comunes como realizar gráficos o calcular la media aritmética de un arreglo.

- Un arreglo permite almacenar una lista de valores. Podemos pensar que es el equivalente a una matriz en álgebra lineal.

````python
# Este bloque importa numpy (para usar arreglos)
# y matplotlib para realizar gráficos
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Este bloque crea un arreglo de 30 valores entre 0 y 1
valores = np.linspace(0, 1, 30)
````

````python
# Este bloque calcula la media aritmética del arreglo valores
# y lo imprime
print(np.mean(valores))
````

````python
# Este bloque
%matplotlib notebook
plt.plot(valores, valores**2);
````

Note que la celda anterior usa el *comando mágico*

```python
%matplotlib notebook
```

que permite tener gráficos interactivos dentro de los notebooks.

También, note que la última instrucción tiene un punto y coma (;) al final. Esto
es para evitar que se muestre el resultado de evaluación de la celda.

# Búsqueda de raíces en Python

```python
import numpy as np
from scipy.optimize import bisect, newton, fsolve
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
def f(x):
    return x**2 + 10*np.sin(x)
```

```python
x = np.linspace(-8, 8, 1000)
y = f(x)
```

```python
plt.plot(x, y)
plt.ylim(-10, 50)
```

```python
fsolve(f, 1)
```

```python
fsolve(f, -3)
```

```python
newton(f, 1)
```

```python
newton(f, 3)
```

```python
bisect(f, -1, 2)
```

```python
bisect(f, -4, -1)
```

```python
fsolve(f, range(-5, 5))
```
